## BRAINSTORM - Como as taxas de vacinao e mortalidade variam entre os diferentes continentes?

```python
deaths = pd.read_csv("/content/drive/MyDrive/Ada/1015/COVID DEATHS.csv")
vaccinations = pd.read_csv("/content/drive/MyDrive/Ada/1015/COVID_VACCINATIONS.csv")
```

```python
dataset = deaths.merge(vaccinations, how="left", on=['continent', 'date', 'iso_code', 'location'])
dataset.sample(10)
```

```python
dataset[dataset['location'] == "Indonesia"]["population"].describe()
```

```python
dataset[dataset['location'] == "Indonesia"]["total_deaths"].describe()
```

```python
grouped_country_df = dataset.groupby(by=['continent', 'location']).agg({
    "total_deaths": "max",
    "population": "last",
    "people_vaccinated": "max"
}).reset_index()
grouped_country_df.head()
```

```python
grouped_df = grouped_country_df.groupby(
    'continent').sum(numeric_only=True).reset_index()
```

```python
grouped_df.head()
```

```python
grouped_df["tax_vaccination"] = grouped_df["people_vaccinated"] / grouped_df["population"]
grouped_df["tax_deaths"] = grouped_df["total_deaths"] / grouped_df["population"]
grouped_df.describe()
```
